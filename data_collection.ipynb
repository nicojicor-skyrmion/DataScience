# Data Collection

In this notebook, we demonstrate an expample of data collection. The data consists of food recipes (cuisines and ingredientes)

```python
import pandas as pd
# URL donde estan los datos almacenados
# En este caso, en un archivo tipo .csv
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-" \
        "DS0103EN-SkillsNetwork/labs/Module%202/recipes.csv"
# Cargar datos
recipes = pd.read_csv(path)
```

```python
# Show the first few rows
recipes.head()
```

```python
# Get the dimensions of the dataframe
recipes.shape
```

So our dataset consists of **57,691 recipes**. Each row represents a recipe, and for each recipe, the corresponding cuisine is documented as well as whether **384 ingredients** exist in the recipe or not, beggining with almond and ending with zucchini. Ingredients are listed in alphabetical order.
