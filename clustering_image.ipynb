```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
from google.colab import files
```

```python
foto = files.upload()
```

```python
nombre_imagen = list(foto.keys())[0]
imagen = cv2.imread(nombre_imagen)
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
pixeles = imagen.reshape((-1, 3))
pixeles = np.float32(pixeles)
```

```python
modelo = KMeans(n_clusters=5)
modelo.fit(pixeles)
```

```python
etiquetas = modelo.predict(pixeles)
centroides = np.uint8(modelo.cluster_centers_)
```

```python
imagen_segmentada = centroides[etiquetas.flatten()]
imagen_segmentada = imagen_segmentada.reshape(imagen.shape)
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(imagen)
plt.title("NICO<3")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(imagen_segmentada)
plt.title(f"NICO EN 5 COLORES FILTRO 4K (K = 5)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
